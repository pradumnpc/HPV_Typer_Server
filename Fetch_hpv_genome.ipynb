## Fetch Complete genome from the query string which contains only lineage information

### Import Module

```python
import re
import requests
from Bio import SeqIO
from Bio import Entrez
```

### Fetch Data  Genome 

```python
c = 0   
file1 = open("HPV_Genome.fasta", "w") # Open file which contains the complete genome of HPV16 virus which have a lineage information 
set1 =set()     # empty set for fetching only unique acession ID 
Entrez.email = 'pradumnpc@gmail.com'

# Using NCBI eutils fetch the all id which contains the complete genome for HPV 16 virus 

x=requests.get('https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=nuccore&term=complete genome[Title] AND "Human papillomavirus type 16"[Organism]&retmax=700')
m=x.text  # URL request into text format for further processiong 
reg=re.finditer('<Id>(\d+)</Id>',m)  # regular expression which fetch only accesion id from given html page in list format 
i=0
 # declare a list which contains the different lineage annotation for same lineage after we itereate all this list to find out exact match for the lineage
A1 = ['genotype: A1','lineage A sublineage A1 (European, E)','lineage A1'] 
A2 = ['lineage A sublineage A2 (European, E)','lineage A2','genotype: A2']  
D3 = ['genotype: D3','lineage D3']

for i in reg:  # itereate loop on all accesion id 
    id=i.group(1)  # get the only accesion ID 
    print(c)
    c+=1
    # url wwhich fetch the genebank file of perticular accesion ID 
    
    url=requests.get('https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=nucleotide&id={}&rettype=gb'.format(id))
    content=url.text  # convert the url request into text format 
    lineage = re.search(r'/note="(.*)"',content)  # regular expression for fetching lineage info of perticular genebank ID 
    if lineage:   #  condition true only when get the lineage information is fetch 
        set1.add(lineage.group(1))
        cond = lineage.group(1) 
#         print(cond)
        for j in A1:  # loop over the A1 list which contains the different lineage annotation for A1 lineage 
            if j == cond:  
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text') # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')  
                print(rec.id)
                sequence = str(rec.seq) 
                file1.write('>A1_{}_l1\n'.format(rec.id)+sequence +'\n')

        for j in A2:
            if j == cond:
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>A2_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage A3':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>A3_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage A4':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>A4_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage B1':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>B1_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage B2':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>B2_{}_l1\n'.format(rec.id)+sequence +'\n') 
        if cond == 'lineage B3':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>B3_{}_l1\n'.format(rec.id)+sequence +'\n')  
        if cond == 'lineage B4':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>B4_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage C1':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>C1_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage C3':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>C3_{}_l1\n'.format(rec.id)+sequence +'\n')   
        if cond == 'lineage C4':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>C4_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage D1':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>D1_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage D2':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>D2_{}_l1\n'.format(rec.id)+sequence +'\n')
        for j in D3:
            if cond == j:
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>D3_{}_l1\n'.format(rec.id)+sequence +'\n')
        if cond == 'lineage D4':
                gb_acc1 = Entrez.efetch(db='nucleotide', id=id, rettype='gb', retmode='text')   # using biopython fetch genebank which contains the linaege 
                rec = SeqIO.read(gb_acc1, 'genbank')
                print(rec.id)
                sequence = str(rec.seq)
                file1.write('>D4_{}_l1\n'.format(rec.id)+sequence +'\n')

file1.close()            
```
